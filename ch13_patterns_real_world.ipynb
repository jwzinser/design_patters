## GoF Pattern Classification

```python
patterns = {
    "Creational": [
        "Abstract Factory", "Builder", "Factory Method",
        "Prototype", "Singleton",
    ],
    "Structural": [
        "Adapter", "Bridge", "Composite", "Decorator",
        "Facade", "Flyweight", "Proxy",
    ],
    "Behavioral": [
        "Chain of Responsibility", "Command", "Interpreter",
        "Iterator", "Mediator", "Memento", "Observer",
        "State", "Strategy", "Template Method", "Visitor",
    ],
}

for family, names in patterns.items():
    print(f"\n{'─'*40}")
    print(f" {family} ({len(names)} patterns)")
    print(f"{'─'*40}")
    for n in names:
        print(f"  • {n}")
```

## Anti-Patterns

Anti-patterns are recurring solutions that seem good but cause more harm than they solve.

```python
anti_patterns = [
    {
        "name": "God Object / Big Ball of Mud",
        "symptom": "One class does everything — thousands of lines, dozens of responsibilities.",
        "fix": "Apply SRP. Extract responsibilities into focused collaborating classes.",
    },
    {
        "name": "Golden Hammer",
        "symptom": "Using a familiar pattern/tool for every problem regardless of fit.",
        "fix": "Learn multiple tools. Always match the tool to the actual problem.",
    },
    {
        "name": "Premature Optimization",
        "symptom": "Sacrificing readability for performance before profiling.",
        "fix": "Profile first. Optimize the measured bottleneck, not imagined ones.",
    },
    {
        "name": "Spaghetti Code",
        "symptom": "Tangled control flow, no clear structure, hard to trace.",
        "fix": "Refactor into methods/classes. Apply patterns where they fit.",
    },
    {
        "name": "Copy-Paste Programming",
        "symptom": "Same logic duplicated in many places.",
        "fix": "DRY — Don't Repeat Yourself. Extract to a shared function or class.",
    },
    {
        "name": "Overuse of Singleton",
        "symptom": "Global state everywhere, tests interfere with each other.",
        "fix": "Prefer dependency injection so collaborators are explicit.",
    },
]

for ap in anti_patterns:
    print(f"\n❌ {ap['name']}")
    print(f"   Symptom: {ap['symptom']}")
    print(f"   Fix:     {ap['fix']}")
```

## The "How to choose a pattern" decision guide

```python
decision_guide = {
    "I need to create objects without specifying exact class":          "Factory Method / Abstract Factory",
    "I need exactly one instance of a class":                          "Singleton (or module-level object)",
    "I need to add responsibilities to objects without subclassing":   "Decorator",
    "I need to convert an incompatible interface":                     "Adapter",
    "I need to simplify a complex subsystem":                          "Facade",
    "I need objects to notify each other of changes":                  "Observer",
    "I need to swap algorithms at runtime":                            "Strategy",
    "I need to encapsulate a request as an object":                    "Command",
    "I need to define an algorithm skeleton, defer steps":             "Template Method",
    "I need object to change behavior based on state":                 "State",
    "I need to traverse a collection uniformly":                       "Iterator",
    "I need a tree of objects treated uniformly":                      "Composite",
    "I need to control access to an object":                           "Proxy",
}

print(f"{'Problem':<55} → Pattern")
print("-" * 80)
for problem, solution in decision_guide.items():
    print(f"{problem:<55} → {solution}")
```

## OO Principles summary — the full list

```python
principles = [
    ("Encapsulate what varies",
     "Identify the aspects that vary and separate them from what stays the same."),
    ("Program to interfaces, not implementations",
     "Depend on abstractions. Don't depend on concrete classes."),
    ("Favor composition over inheritance",
     "HAS-A is often better than IS-A. Combine behaviors at runtime."),
    ("Strive for loosely coupled designs",
     "Minimize dependencies between objects that interact."),
    ("Open/Closed Principle",
     "Classes should be open for extension, closed for modification."),
    ("Dependency Inversion Principle",
     "Depend on abstractions. High-level modules should not depend on low-level modules."),
    ("Principle of Least Knowledge (Law of Demeter)",
     "Talk only to your immediate friends. Don't chain method calls."),
    ("Hollywood Principle",
     "Don't call us, we'll call you. High-level components call low-level ones, not vice versa."),
    ("Single Responsibility Principle",
     "A class should have only one reason to change."),
]

for i, (name, desc) in enumerate(principles, 1):
    print(f"\n{i}. {name}")
    print(f"   {desc}")
```

## Pattern recognition practice

```python
quiz = [
    ("Django signals (post_save, pre_delete)", "Observer"),
    ("Python sorted(key=lambda x: x.age)", "Strategy"),
    ("@functools.lru_cache", "Proxy (Caching)"),
    ("Python @property getter/setter", "Proxy (Protection)"),
    ("io.BufferedReader(io.FileIO(...))", "Decorator"),
    ("pathlib.Path('/a') / 'b' / 'c'", "Composite"),
    ("requests.get() hiding socket/http complexity", "Facade"),
    ("unittest.TestCase setUp/tearDown hooks", "Template Method"),
    ("logging.getLogger() returns same logger instance", "Singleton"),
    ("itertools module", "Iterator"),
    ("contextlib.contextmanager", "Template Method / Decorator"),
]

print(f"{'Example':<50} {'Pattern'}")
print("-" * 70)
for example, pattern in quiz:
    print(f"{example:<50} {pattern}")
```

## Interview cheat-sheet

| Question | Answer |
|---|---|
| What is a pattern? | A named, reusable solution to a recurring design problem in a specific context. |
| What is an anti-pattern? | A commonly used but counterproductive solution. |
| Should you always use a pattern? | No. Don't force patterns — simplest solution that works is best. |
| What's a pattern language? | A system of related patterns that work together to solve problems at multiple scales. |
